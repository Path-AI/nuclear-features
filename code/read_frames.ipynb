# Annotation loading for the nuclei segmentation model

We're able to share nucleus annotations for train, validation, test, and OOD-2 datasets; as well as metadata associated with each.

```python
# imports and functions in general
from analysis import dataset_utils as du

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

Read in frames from metadata Excel file.

```python
frames_meta = pd.read_excel("../data/annotations/nuclei_annotations.xlsx", index_col="Slide IDs")
frame_ids = frames_meta["Frame IDs"].values

frames_loaded = {}
for frame_id in frame_ids:
    frames_loaded[frame_id] = du.Frame(frame_id, frame_directory="../data/annotations/")
```
